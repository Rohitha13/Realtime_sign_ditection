# data collection  when mode is testing - images goto test data folder and vice versa

```python
# Importing the Libraries Required

import cv2
import numpy as np
import os

# Creating and Collecting Training Data

mode = 'testingData'
directory = 'dataSet2/' + mode + '/'
minValue = 35

capture = cv2.VideoCapture(0)
interrupt = -1

while True:
    _, frame = capture.read()

    # Simulating mirror Image

    frame = cv2.flip(frame, 1)

    # Getting count of existing images

    count = {  #change this according to the order you want
#                 'zero': len(os.listdir(directory+"/0")),

                '1': len(os.listdir(directory+"/1")),
                '2': len(os.listdir(directory+"/2")),
                '3': len(os.listdir(directory+"/3")),
                '4': len(os.listdir(directory+"/4")),
                '5': len(os.listdir(directory+"/5")),
                
                
    }

    # Printing the count of each set on the screen
    
#     cv2.putText(frame, "ZERO : " +str(count['zero']), (10, 60), cv2.FONT_HERSHEY_PLAIN, 1, (0,255,255), 1)
    
    cv2.putText(frame, "1 : " +str(count['1']), (10, 70), cv2.FONT_HERSHEY_PLAIN, 1, (0,255,255), 1)
    cv2.putText(frame, "2 : " +str(count['2']), (10, 80), cv2.FONT_HERSHEY_PLAIN, 1, (0,255,255), 1)
    cv2.putText(frame, "3 : " +str(count['3']), (10, 90), cv2.FONT_HERSHEY_PLAIN, 1, (0,255,255), 1)
    cv2.putText(frame, "4 : " +str(count['4']), (10, 100), cv2.FONT_HERSHEY_PLAIN, 1, (0,255,255), 1)
    cv2.putText(frame, "5 : " +str(count['5']), (10, 110), cv2.FONT_HERSHEY_PLAIN, 1, (0,255,255), 1)



    # Coordinates of the ROI
    
    x1 = int(0.5*frame.shape[1])
    y1 = 10
    x2 = frame.shape[1]-10
    y2 = int(0.5*frame.shape[1])

    # Drawing the ROI
    # The increment/decrement by 1 is to compensate for the bounding box
    
    cv2.rectangle(frame, (x1-1, y1-1), (x2+1, y2+1), (255,0,0) ,1)
    
    # Extracting the ROI
    
    roi = frame[y1:y2, x1:x2]

    cv2.imshow("Frame", frame)
    
    # Image Processing

    roi = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
    _, roi = cv2.threshold(roi, 120, 255, cv2.THRESH_BINARY)
    cv2.imshow("ROI", roi)

    # Data Collection

    interrupt = cv2.waitKey(10)
    if interrupt & 0xFF == 27: 
        # esc key
        break
#     if interrupt & 0xFF == ord('0'):
#         cv2.imwrite(directory+'0/'+str(count['zero'])+'.jpg', roi)

    if interrupt & 0xFF == ord('1'):
        cv2.imwrite(directory+'1/'+str(count['1'])+'.jpg', roi)
    
    if interrupt & 0xFF == ord('2'):
        cv2.imwrite(directory+'2/'+str(count['2'])+'.jpg', roi)
    
    if interrupt & 0xFF == ord('3'):
        cv2.imwrite(directory+'3/'+str(count['3'])+'.jpg', roi)
    
    if interrupt & 0xFF == ord('4'):
        cv2.imwrite(directory+'4/'+str(count['4'])+'.jpg', roi)
    
    if interrupt & 0xFF == ord('5'):
        cv2.imwrite(directory+'5/'+str(count['5'])+'.jpg', roi)
    
    interrupt = cv2.waitKey(1)
    if interrupt & 0xFF == ord('q'): # esc key
        break
  
    
capture.release()
cv2.destroyAllWindows()
```

# model train

```python
# Importing the Keras libraries and packages
from keras.models import Sequential
from keras.layers import Convolution2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten , Dropout
from keras.layers import Dense

# Step 1 - Building the CNN

# Initializing the CNN
classifier = Sequential()

# First convolution layer and pooling
classifier.add(Convolution2D(32, (3, 3), input_shape=(64, 64, 1), activation='relu'))
classifier.add(MaxPooling2D(pool_size=(2, 2)))


classifier.add(Dropout(0.25))
# Here 0.25 means we are avoiding 25% of the neurons to avoid overfitting of the data.

# Second convolution layer and pooling
classifier.add(Convolution2D(64, (3, 3), activation='relu'))
# input_shape is going to be the pooled feature maps from the previous convolution layer
classifier.add(MaxPooling2D(pool_size=(2, 2)))

# 3rd
classifier.add(Convolution2D(128, (3, 3), activation='relu'))
# input_shape is going to be the pooled feature maps from the previous convolution layer
classifier.add(MaxPooling2D(pool_size=(2, 2)))

# 4th
classifier.add(Convolution2D(256, (3, 3), activation='relu'))
# input_shape is going to be the pooled feature maps from the previous convolution layer
classifier.add(MaxPooling2D(pool_size=(2, 2)))


# Flattening the layers
classifier.add(Flatten())

# Adding a fully connected layer
classifier.add(Dense(units=128, activation='relu'))
classifier.add(Dense(units=11, activation='softmax')) # softmax for more than 2 categories

# Compiling the CNN
classifier.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy']) # categorical_crossentropy for more than 2


# Step 2 - Preparing the train/test data and training the model
from keras.preprocessing.image import ImageDataGenerator
train_datagen = ImageDataGenerator(
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True)

test_datagen = ImageDataGenerator(rescale=1./255)

training_set = train_datagen.flow_from_directory(r'C:\Users\rohitha\Downloads\Sign_Language_Recognition_with_GUI-main\Sign_Language_Recognition_with_GUI-main\dataSet2\trainingData',
                                                 target_size=(64, 64),
                                                 batch_size=10,
                                                 color_mode='grayscale',
                                                 class_mode='categorical')

test_set = test_datagen.flow_from_directory(r'C:\Users\rohitha\Downloads\Sign_Language_Recognition_with_GUI-main\Sign_Language_Recognition_with_GUI-main\dataSet2\testingData',
                                            target_size=(64, 64),
                                            batch_size=5,
                                            color_mode='grayscale',
                                            class_mode='categorical') 
classifier.fit(
        training_set,
        #steps_per_epoch=3300, # No of images in training set
        #steps_per_epoch=820,
        epochs=5,
        validation_data=test_set)



        #validation_steps=110) # No of images in test set
        #validation_steps=715) # No of images in test set


print("done")
# Saving the model
model_json = classifier.to_json()
with open("mew.json", "w") as json_file:
    json_file.write(model_json)
classifier.save_weights('mew.h5')




```
